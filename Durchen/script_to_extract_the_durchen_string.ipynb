```python
from pathlib import Path
import re
import yaml
```

```python
def extracting_durchen_string(poti_content, start, end): 
    durchen_string = ""
    pages = get_pages(poti_content)
    durchen_string = "".join(pages[(start-1):(end-1)])
    return durchen_string
```

```python
def get_pages(poti_content):
    result = []
    pg_text = ""
    pages = re.split(f"(\[\d+\w+\])", poti_content)
    for i, page in enumerate(pages[1:],1):
        if i % 2 == 0:
            pg_text += page
        else:
            pg_text += page
            result.append(pg_text)
            pg_text = ""
    return result
```

```python
def poti_update(result,poti_num):
    print("end of durchen extraction")
    
    Path(f"./durchen_string/{poti_num}").write_text(result, encoding = 'utf-8')
    print(f'{poti_num} Update complete .....')
```

```python
def get_potis_and_yaml():
    t2d = list(Path(f'./tengyur_dictionary').iterdir())
    t2d.sort()
    all_potis = list(Path(f'./clean_tengyur_google_text').iterdir())
    all_potis.sort()
    
    for poti_path, t2d_path in zip(all_potis, t2d):
        t2d_content = t2d_path.read_text(encoding = 'utf-8')
        t2d_content = yaml.safe_load(t2d_content)
        print(f'yaml file {t2d_path.name} processing')
        poti_content = poti_path.read_text(encoding = 'utf-8')
        poti_num = poti_path.name
        print(f'poti file {poti_num} processing')
        result = ""
        for text_id, t2d_text in t2d_content.items():
            pg_start = int(t2d_text['start'])
            pg_end = int(t2d_text['end'])
            durchen_string = extracting_durchen_string(poti_content, pg_start, pg_end)
            result += durchen_string
            durchen_string = ""
        poti_update(result,poti_num)
        print("poti update works \n")
```

```python
get_potis_and_yaml()
```
